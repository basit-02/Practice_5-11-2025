Step 0: Imports Essential Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

plt.style.use('ggplot')
```

Step 1: Uploading Dataset and understand it

```python
df = pd.read_csv('coaster_db.csv')

print("\nOriginal Shape of the dataset:\n", df.shape)
print("\nSee Starting 5 records:\n", df.head())
print("\nCalculating Missing Values:\n", df.isna().sum())
```

Step 2: Remove Columns with more than 80% Missing

```python
missing_percent = (df.isna().sum() / len(df)) * 100
cols_to_drop = missing_percent[missing_percent > 80].index

print("\nDropping columns (missing > 80%):")
print(cols_to_drop)

df = df.drop(columns=cols_to_drop)
print("Shape after dropping Columns :", df.shape)
```
